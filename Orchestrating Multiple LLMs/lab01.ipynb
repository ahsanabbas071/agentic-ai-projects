```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")

if openai_api_key:
    print(f"api key exists and starts with {openai_api_key[:8]}")
else:
    print("api key not found. troubleshoot")
```

```python
request = "Tell me the fastest way to earn proper money in 2026 and onwards from anywhere in the world. What skills are high in demand but there is shortage of those skills. Tell me the skills that take less than few months to learn. Only answer with skills name and in bullet points"

messages = [{ "role" : "user", "content": request}]

```

```python
openai = OpenAI()
```

```python
response = openai.chat.completions.create(
    model = "gpt-5-mini",
    
    messages = messages
)

skills = response.choices[0].message.content
display(Markdown(skills))

```
